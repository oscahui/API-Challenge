```python
from api_keys import (weather_api_key, g_key)
import numpy as np
import pandas as pd
import requests
import time
from scipy.stats import linregress
from matplotlib import pyplot as plt
from citipy import citipy
from pprint import pprint
import random
```

## Generate Cities List

```python
import random
randomlat=[]
randomlng=[]
city_l=[]
country_l=[]
no=2000
for i in range(no):
    lat = random.uniform(-90,90)
    lng = random.uniform(-180,180)
    lat=round(lat,5)
    lng=round(lng,5)
    randomlat.append(lat)
    randomlng.append(lng)    
    city = citipy.nearest_city(lat, lng)
    cities = city.city_name
    countries = city.country_code    
    city_l.append(cities)
    country_l.append(countries)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={weather_api_key}&units={units}&q="
lat_1=[]
temp=[]
max_temp=[]
min_temp=[]
humid=[]
wind=[]
cloud=[]
n_lat=[]
n_lon=[]
for city_list in city_l:
    response = requests.get(query_url + city_list).json()
    try:
        max_temp.append(response['main']['temp_max'])
        min_temp.append(response['main']['temp_min'])    
        temp.append(response['main']['temp'])     
        humid.append(response['main']['humidity'])        
        cloud.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        n_lat.append(response['coord']['lat'])
        n_lon.append(response['coord']['lon'])
        print(f"{city_list}{response['id']} found! Appending stats")
    except:
        max_temp.append(0)
        min_temp.append(0)    
        temp.append(0)        
        humid.append(0)  
        cloud.append(0)  
        wind.append(0)  
        n_lat.append(0)  
        n_lon.append(0)  
        print("error")
        pass
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#my_dict={'Country':country_l,'City':city_l,'Lat':n_lat,'Lnt':n_lon,'Temperature':temp,'Max_Temp':max_temp,'Humidity':humid,'Cloudiness':cloud,'Wind_Speed':wind}
#df= pd.DataFrame.from_dict(my_dict, orient='index')
#df=df.transpose()

df = pd.DataFrame({'Country':country_l,'City':city_l,'Lat':n_lat,'Lnt':n_lon,'Temperature':temp,'Max_Temp':max_temp,'Humidity':humid,'Cloudiness':cloud,'Wind_Speed':wind})
df=df.drop_duplicates(subset=['City'])
drop_error_df=df.drop_duplicates(subset=['Lat','Lnt'],keep=False)
drop_error_df.to_csv('citydata.csv', index=False)
df
```

## Linear Regression

```python
north_clean_city_data=drop_error_df.loc[df['Lat'] >= 0]
south_clean_city_data=drop_error_df.loc[df['Lat'] <= 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
N_lat=north_clean_city_data['Lat']
N_temp=north_clean_city_data['Temperature']


vc_slope, vc_int, vc_r, vc_p, vc_std_err = linregress(N_lat, N_temp)
vc_fit = vc_slope * N_temp + vc_int
plt.scatter(N_lat,N_temp)
plt.plot(N_temp,vc_fit,"--")
plt.title("Nothern Hemisphere - Temp vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temp")
plt.savefig('../Output/N-Temp v Lat.png')
plt.show()
```
